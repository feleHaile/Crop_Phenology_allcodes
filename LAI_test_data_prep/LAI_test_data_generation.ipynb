```python
import pandas as pd
import numpy as np
```

```python
crop_info=pd.read_excel("../Nebraska_Planting&Harvesting.xlsx",sheetname='crop_info')
lai=pd.read_excel("Ameriflux Final Crop Data Set - 2001 to 2016.xlsx",skiprows=[0,1,2,3,4])
lai=lai.drop([0])
lai = lai.drop(['Sample Date','Canopy Height'], 1)
lai = lai[np.isfinite(lai['Year'])]
lai[['Year','Site','Sample\n(DOY)']]=lai[['Year','Site','Sample\n(DOY)']].astype(int)
lai=pd.merge(lai, crop_info,  how='left', left_on=['Year','Site'], right_on = ['Year','Site'])
lai
```

```python
temp=pd.read_csv("meads_bands/ne1_corn.csv").drop('Unnamed: 0',1)
temp['crop']=1
temp['site']=1
meads_data=temp

temp=pd.read_csv("meads_bands/ne1_soy.csv").drop('Unnamed: 0',1)
temp['crop']=2
temp['site']=1
meads_data=meads_data.append(temp,ignore_index=True)

temp=pd.read_csv("meads_bands/ne2_corn.csv").drop('Unnamed: 0',1)
temp['crop']=1
temp['site']=2
meads_data=meads_data.append(temp,ignore_index=True)

temp=pd.read_csv("meads_bands/ne2_soy.csv").drop('Unnamed: 0',1)
temp['crop']=2
temp['site']=2
meads_data=meads_data.append(temp,ignore_index=True)

temp=pd.read_csv("meads_bands/ne3_corn.csv").drop('Unnamed: 0',1)
temp['crop']=1
temp['site']=3
meads_data=meads_data.append(temp,ignore_index=True)

temp=pd.read_csv("meads_bands/ne3_soy.csv").drop('Unnamed: 0',1)
temp['crop']=2
temp['site']=3
meads_data=meads_data.append(temp,ignore_index=True)

meads_data
```

```python
from functools import reduce
def return_index(x):
    yr=x[0]
    col=2*(yr-2000)-1
    sitee=x[1]
    cropp=x[4]
    doyy=x[2]
    idx=reduce(np.intersect1d,(np.where(meads_data['site'] == sitee)[0],np.where(meads_data['crop'] == cropp)[0],np.where(meads_data['doy'] == doyy)[0]))[0]
    return meads_data.loc[idx][col],meads_data.loc[idx][col+1]
lai[['b1','b2']]=lai.apply(return_index,axis=1).apply(pd.Series)
#lai[['b1','b2']]=lai['bands'].apply(pd.Series)
lai
```
