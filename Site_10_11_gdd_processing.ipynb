```python
import pandas as pd
import datetime

def doy(r,col_name="Date"):
    return datetime.datetime(int(r[col_name][:4]), int(r[col_name][5:7]), int(r[col_name][8:10])).timetuple().tm_yday

r = pd.read_excel("Ames_Brooks_Data.xlsx",sheetname="Met_Data")
r=r[['TimeStamp','Air Temp (◦F)']]
r.iloc[:,0]=r.applymap(str).iloc[:,0]
r['Date'] = r.TimeStamp.str[:11]
r['Year'] = r.TimeStamp.str[:4]
r['DOY'] = pd.Series.to_frame(r.apply(doy,axis=1))
r=r.loc[r['Year'].isin(['2002','2003','2004','2005'])]
r['temp(C)']=5*(r['Air Temp (◦F)']-32)/9
r=r[['Date','Year','DOY','temp(C)']]
```

```python
df=r.groupby('Date')['temp(C)'].max().reset_index()
df.columns=["Date",'tmax']
df['tmin']=r.groupby('Date')['temp(C)'].min().reset_index()['temp(C)']
df['Year'] = df.Date.str[:4]
df['DOY'] = pd.Series.to_frame(df.apply(doy,axis=1))
df
```

```python
daily = pd.read_excel("Ames_Brooks_Data.xlsx",sheetname="Met_Data2")
daily=daily[['Date','High Air Temp (°C)','Low Air Temp (°C)']]
daily.columns=['Date','tmax','tmin']
daily.loc[:,'Date']=daily.applymap(str).loc[:,'Date']
daily['Date'] = daily.Date.str[:11]
daily['Year'] = daily.Date.str[:4]
daily['DOY'] = pd.Series.to_frame(daily.apply(doy,axis=1))
daily
```

```python
temp_data=pd.concat([df,daily])
temp_data.index=range(1,1+len(temp_data.index))
temp_data
```

```python
import pandas as pd
temp_data=pd.read_pickle("Temperature_data_2002to2013_Site_10_11")
temp_data['tmin_corn'] = temp_data.tmin.clip(8.0,30.0)
temp_data['tmax_corn'] = temp_data.tmax.clip(8.0,30.0)
temp_data['tmin_soy'] = temp_data.tmin.clip(10.0,25.0)
temp_data['tmax_soy'] = temp_data.tmax.clip(10.0,25.0)
temp_data['tavg_corn']=0.5*(temp_data.tmin_corn+temp_data.tmax_corn)
temp_data['tavg_soy']=0.5*(temp_data.tmin_soy+temp_data.tmax_soy)
temp_data['gdd_corn']=temp_data['tavg_corn']-8.0
temp_data['gdd_soy']=temp_data['tavg_soy']-10.0
temp_data
temp_data.to_pickle("GDD_Temperature_data_2002to2013_Site_10_11")

#pd.concat([temp_data, pd.concat([df4], keys=["tmin"])])
```

```python
gdd_corn_green=340
gdd_soy_green=400
gdd_corn_sen=150
gdd_soy_sen=260

g10=[159,189,181,162,186,169,183,159,158,163,166,169]#,185,164]
s10=[248,256,269,240,257,256,252,271,247,255,233,266]#,263,257]
g11=[163,188,161,159,162,172,181,159,157,165,166,199]#,161,172]
s11=[218,253,271,242,243,254,251,224,247,258,246,265]#,263,260]

def gdd_backcalculation(green,sen,gdd_corn_green,gdd_soy_green,gdd_corn_sen,gdd_soy_sen):
    p_corn=[]
    p_soy=[]
    h_corn=[]
    h_soy=[]
    for i in range(len(green)) :
        yr=str(2002+i)
        g=green[i]
        s=sen[i]

        t=temp_data.loc[(temp_data.Year==yr) & (temp_data.DOY<=g)]
        t.index=range(1,1+len(t))
        t['gdd_corn_cum']=t.loc[::-1, 'gdd_corn'].cumsum()[::-1]
        t['gdd_soy_cum']=t.loc[::-1, 'gdd_soy'].cumsum()[::-1]
        c_cum=t['gdd_corn_cum']
        s_cum=t['gdd_soy_cum']
        corn=c_cum[c_cum<gdd_corn_green]
        soy=s_cum[s_cum<gdd_soy_green]
        p_corn.append(corn.index[0]-1 if(corn.index[0]>1) else 100)
        p_soy.append(soy.index[0]-1 if(soy.index[0]>1) else 100)

        t=temp_data.loc[(temp_data.Year==yr) & (temp_data.DOY>=s)]
        t.index=range(s,s+len(t))
        t['gdd_corn_cum']=t.loc[:,'gdd_corn'].cumsum()
        t['gdd_soy_cum']=t.loc[:,'gdd_soy'].cumsum()
        c_cum=t['gdd_corn_cum']
        s_cum=t['gdd_soy_cum']
        corn=c_cum[c_cum>gdd_corn_sen]
        soy=s_cum[s_cum>gdd_soy_sen]
        h_corn.append(corn.index[0] if len(corn.index>0) else 320)
        h_soy.append(soy.index[0] if len(soy.index>0) else 320)
    return (p_corn,p_soy,h_corn,h_soy)
p10_corn,p10_soy,h10_corn,h10_soy=gdd_backcalculation(g10,s10,gdd_corn_green,gdd_soy_green,gdd_corn_sen,gdd_soy_sen)
p11_corn,p11_soy,h11_corn,h11_soy=gdd_backcalculation(g11,s11,gdd_corn_green,gdd_soy_green,gdd_corn_sen,gdd_soy_sen)
```
